```python
import requests, bs4
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://coinmarketcap.com/all/views/all/'
response = requests.get(url)
status = response.status_code
if status == 200:
    page = response.text
    soup = bs(page)
else:
    print(f'Oops! Received status code{status}')

soup
```

```python
import re
list1 = []

for div in soup.find_all(class_= 'sc-131di3y-0 cLgOOr'):
    title_tag = div.find(class_='cmc-link')
    title = title_tag
    price = title
    list1.append({price})
    
list1


```

```python
list2 = []

for i in range(0, len(list1)):
    word = list1[i]
    word2 = list(word)
    list2.append(word2[0].text)
    
list2

```

```python


list_ticker = []

for div in soup.find_all(class_= 'cmc-table__column-name--symbol cmc-link'):
    ticker = div.next
    list_ticker.append({ticker})
    

list_ticker

ticker = list_ticker[0]
ticker2 = list(ticker)
ticker2[0]

list_ticker2 = []

for i in range(0, len(list_ticker)):
    ticker = list_ticker[i]
    ticker2 = list(ticker)
    list_ticker2.append(ticker2[0])

list_ticker2
```

```python
ticker_price = dict(zip(list_ticker2, list2))
ticker_price
```
