```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get('https://coinmarketcap.com/all/views/all/')
time.sleep(1)

#for i in range(1,21):
    #ticker = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[2]/div[3]/div/table/tbody/tr['+str(i)+']/td[3]/div').text;
    #print(ticker)

#1000 - HNT
#1500 - ONE
#2000 - HNT

#Next page
driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
load_more = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[3]/button').click()
time.sleep(3)
driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
load_more = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[3]/button').click()
time.sleep(3)
driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
load_more = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[3]/button').click()
time.sleep(3)
driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
load_more = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[3]/button').click()
time.sleep(3)

#driver.execute_script('return document.body.scrollHeight')

driver.execute_script("window.scrollTo(0,1500)")
time.sleep(1)
driver.execute_script("window.scrollTo(1500,3000)")
time.sleep(1)
driver.execute_script("window.scrollTo(3000,4500)")
time.sleep(1)
driver.execute_script("window.scrollTo(4500,6000)")
time.sleep(1)
driver.execute_script("window.scrollTo(6000,7500)")
time.sleep(1)
driver.execute_script("window.scrollTo(7500,9000)")
time.sleep(1)
driver.execute_script("window.scrollTo(9000,10500)")
time.sleep(1)
driver.execute_script("window.scrollTo(10500,12000)")
time.sleep(1)
driver.execute_script("window.scrollTo(12000,13500)")
time.sleep(1)
driver.execute_script("window.scrollTo(13500,15000)")
time.sleep(1)
driver.execute_script("window.scrollTo(15000,16500)")
time.sleep(1)
driver.execute_script("window.scrollTo(16500,18000)")
time.sleep(1)
driver.execute_script("window.scrollTo(18000,19500)")
time.sleep(1)
driver.execute_script("window.scrollTo(19500,21000)")
time.sleep(1)
driver.execute_script("window.scrollTo(21000,22500)")
time.sleep(1)
driver.execute_script("window.scrollTo(22500,24000)")
time.sleep(1)
driver.execute_script("window.scrollTo(24000,25500)")
time.sleep(1)
driver.execute_script("window.scrollTo(25500,27000)")
time.sleep(1)
driver.execute_script("window.scrollTo(27000,28500)")
time.sleep(1)
driver.execute_script("window.scrollTo(28500,30000)")
time.sleep(1)
driver.execute_script("window.scrollTo(30000,31500)")
time.sleep(1)
driver.execute_script("window.scrollTo(31500,33000)")
time.sleep(1)
driver.execute_script("window.scrollTo(33000,34500)")
time.sleep(1)
driver.execute_script("window.scrollTo(34500,36000)")
time.sleep(1)
driver.execute_script("window.scrollTo(36000,37500)")
time.sleep(1)
driver.execute_script("window.scrollTo(37500,39000)")
time.sleep(1)
driver.execute_script("window.scrollTo(39000,40500)")
time.sleep(1)
driver.execute_script("window.scrollTo(40500,42000)")
time.sleep(1)
driver.execute_script("window.scrollTo(42000,43500)")
time.sleep(1)
driver.execute_script("window.scrollTo(43500,45000)")
time.sleep(1)
driver.execute_script("window.scrollTo(45000,45253)")

symbol = []
market_cap = []
price = []
circulation = []
volume = []
one_hour = []
twenty_four = []
seven_day = []

for i in range(1, 1001):
    ticker = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[2]/div[3]/div/table/tbody/tr['+str(i)+']/td[3]/div').text;
    symbol.append(ticker)
    marketCap = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[2]/div[3]/div/table/tbody/tr['+str(i)+']/td[4]/p/span[2]').text;
    market_cap.append(marketCap)
    price_= driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[2]/div[3]/div/table/tbody/tr['+str(i)+']/td[5]/div/a/span').text;
    price.append(price_)
    circulation_ = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[2]/div[3]/div/table/tbody/tr['+str(i)+']/td[6]/div').text;
    circulation.append(circulation_)
    volume_ = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[2]/div[3]/div/table/tbody/tr['+str(i)+']/td[7]/a').text;
    volume.append(volume_)
    oneHour = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[2]/div[3]/div/table/tbody/tr['+str(i)+']/td[8]/div').text;
    one_hour.append(oneHour)
    twentyFourHours = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[2]/div[3]/div/table/tbody/tr['+str(i)+']/td[9]/div').text;
    twenty_four.append(twentyFourHours)
    sevenDays = driver.find_element_by_xpath('//*[@id="__next"]/div[1]/div[2]/div/div[1]/div/div[2]/div[3]/div/table/tbody/tr['+str(i)+']/td[10]/div').text;
    seven_day.append(sevenDays)
```
