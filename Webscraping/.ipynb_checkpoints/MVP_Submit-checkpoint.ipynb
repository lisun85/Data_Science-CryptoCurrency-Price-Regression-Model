```python
import statsmodels.api as sm
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv("Data_Crypto2.csv")
df
```

```python
x = df.drop(columns = ['Price', 'Symbol'])
x = sm.add_constant(x)
y = df['Price']
z = df[['Price', '% 1h', '% 24h', '% 7d']]
z2 = df[['Price', 'Market Cap']]
z3 = df[['Price', 'Volume(24h)']]
z4 = df[['Price', 'Circulating Supply']]

model = sm.OLS(y, x, data = df)
crypto_model = model.fit()
crypto_model.summary()
```
